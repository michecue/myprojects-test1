```python
def find_max(numbers):
  max_number = numbers[0]

  for num in numbers:
    if num > max_number:
      max_number = num

  return max_number

numbers = [0,15,1,7,3,-20]
print(find_max(numbers))
```

```python
def max(a, b, c):
  if a > b and a > c:
    return a
  elif a < b and a > c:
    return b
  else:
    return c

num1 = int(input("Enter a number: "))
num2 = int(input("Enter a number: "))
num3 = int(input("Enter a number: "))

max_num = max(num1, num2, num3)
print(max_num, " is the biggest number.")
```
